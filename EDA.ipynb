```python
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize
```

```python
df = pd.read_csv("data/BA_reviews.csv")
```

```python
df.head()
```

```python
# Character counts
df["char_count"] = df["reviews"].map(lambda x: len(x))
# Word counts
df["word_count"] = df["reviews"].map(lambda x: len(x.split()))
# Sentence counts
df["sent_count"] = df["reviews"].map(lambda x: len(sent_tokenize(x)))
# String counts
df["hashtag_count"] = df["reviews"].map(lambda x: x.count("#"))
# Average word length
df["avg_word_len"] = df["reviews"].map(lambda x: np.mean([len(w) for w in str(x).split()]))
# Average sentence length
df["avg_sent_len"] = df["reviews"].map(lambda x: np.mean([len(w.split()) for w in sent_tokenize(x)]))
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(12,10))
sns.histplot(data = df, x ="char_count",kde = True,ax=axs[0][0])
sns.histplot(data = df, x ="word_count",kde = True,ax=axs[0][1])
sns.histplot(data = df, x ="sent_count",kde = True,ax=axs[1][0])
sns.histplot(data = df, x ="avg_word_len",kde = True,ax=axs[1][1])
sns.histplot(data = df, x ="avg_sent_len",kde = True,ax=axs[2][0])
axs[2][1].axis('off')
```
